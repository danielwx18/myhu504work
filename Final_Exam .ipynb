```python
import random
num = random.randint(1,100)
```

```python
def guessNum(num):
    tmp = 50;
    while tmp>=1 and tmp<=100:
        if tmp==num:
            print("Your guess is "+str(tmp)+", which is correct, the number is "+str(num))
            return
        if tmp<num:
            print("Your guess is "+str(tmp)+", which is lower")
            tmp=tmp+1
        else:            
            print("Your guess is "+str(tmp)+", which is higher")
            tmp=tmp-1
```

```python
guessNum(num)
```
