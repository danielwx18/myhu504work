```python
missing_vals = ["9999999", "Miss", "MMM", "999","0.0"]
omit_data1 = pd.read_csv('Certificate_of_Need_Applications__Beginning_1974.csv', na_values = missing_vals)
```

```python
omit_data1 = omit_data1[pd.notnull(omit_data1['Total Project Cost'])]
```

```python
omit_data1.head(10)
```

```python
filter = (pd.to_datetime(omit_data1["Project Status Date"])>='01/01/2014')&(pd.to_datetime(omit_data1["Project Status Date"])<='12/31/2019')
omit_year1 = omit_data1[filter]
```

```python
omit_year1.head(5000)
```

```python
g1 = sns.FacetGrid(omit_year1, col="Facility Type")
g1.map(plt.hist, "Total Project Cost");
```

```python
omit_year1["Project Status Date"] = (pd.to_datetime(omit_year1["Project Status Date"])).dt.year
g2 = sns.FacetGrid(omit_year1, col="Facility Type", hue="Project Status Date")
g2.map(plt.scatter, "Total Project Cost", "Project Status Date", alpha=.7)
g2.add_legend();
```

```python
g3 = sns.FacetGrid(omit_year1, col="Project Status", hue="Project Status Date")
g3.map(plt.scatter, "Total Project Cost", "Project Status Date", alpha=.7)
g3.add_legend();
```
